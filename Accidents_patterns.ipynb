```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from datetime import datetime
```

```python
url = "https://raw.githubusercontent.com/fivethirtyeight/data/master/bad-drivers/bad-drivers.csv"
df = pd.read_csv(url)

print(df.head())

print(df.describe())
```

```python
df.info()
```

```python
print("Missing Values:\n", df.isnull().sum())
```

```python
plt.figure(figsize=(12, 6))
df = df.sort_values("Number of drivers involved in fatal collisions per billion miles", ascending=False)
sns.barplot(y="State", x="Number of drivers involved in fatal collisions per billion miles", data=df, palette="viridis")
plt.title("State-wise Fatal Collisions Per Billion Miles", fontsize=16)
plt.xlabel("Fatal Collisions Per Billion Miles")
plt.ylabel("State")
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(
    x="Percentage Of Drivers Involved In Fatal Collisions Who Were Speeding",
    y="Percentage Of Drivers Involved In Fatal Collisions Who Were Alcohol-Impaired",
    data=df,
    hue="State",
    palette="tab20",
    legend=False
)
plt.title("Speeding vs Alcohol-Impaired Drivers", fontsize=16)
plt.xlabel("Speeding (%)")
plt.ylabel("Alcohol-Impaired (%)")
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.regplot(
    x="Car Insurance Premiums ($)",
    y="Losses incurred by insurance companies for collisions per insured driver ($)",
    data=df,
    scatter_kws={'color': 'purple'},
    line_kws={'color': 'orange'}
)
plt.title("Insurance Premiums vs Losses Incurred", fontsize=16)
plt.xlabel("Car Insurance Premiums ($)")
plt.ylabel("Losses Incurred ($)")
plt.show()
```

```python
plt.figure(figsize=(10, 6))
corr = df.corr()
sns.heatmap(corr, annot=True, cmap="coolwarm", fmt=".2f")
plt.title("Correlation Heatmap", fontsize=16)
plt.show()
```
